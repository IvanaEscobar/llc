```python
from getSTDOUT import *

path_to_STDOUT='data/'; extension='.notides.'; numRuns=4;
```

```python
d_notides={}; d={}
for i in range(numRuns):
    d[i]=getData( (path_to_STDOUT + 'STDOUT.0000.' + str(i+1)) )
    d_notides[i]=getData( (path_to_STDOUT + 'STDOUT.0000' + extension + str(i+1)) )
```

```python
# Variable names of interest
var1 = 'advcfl wvel max'
var2 = 'dynstat sst mean'
var_list=[var1,var2]
```

```python
#Plotting
import matplotlib.pyplot as plt
from numpy import array

# if you want to plot operations on the vals (ie. a difference between two instances) you need to make an array
# else, plot away
r=4;c=3;
for v in var_list:
    plt.figure()
    plt.suptitle(v)
    for i in range(r):
        notide=array( d_notides[i].getVals( v ) )
        tide=array( d[i].getVals( v ) )
        
        plt.subplot(r,c, c*(i+1)-2 )
        plt.plot( notide )
        if i==0:
            plt.title( 'No Tides' )
            
        plt.subplot(r,c, c*(i+1)-1 )
        plt.plot( tide )
        if i==0:
            plt.title( 'Tides' )
            
        plt.subplot(r,c, c*(i+1) )
        if i==0 or i==1: #currently working on the 3rd run for viz
            plt.plot( tide-notide, 'r-' )
        if i==0:
            plt.title( 'Diff' )
    plt.savefig( (v + '.png'), dpi=1200 )
```
